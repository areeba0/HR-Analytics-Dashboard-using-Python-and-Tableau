```python
import pandas as pd
from faker import Faker
import random

fake = Faker()
num_records = 1000

job_roles = ['Software Engineer', 'Data Scientist', 'Product Manager', 'Sales Executive', 'HR Manager']
departments = ['IT', 'Sales', 'HR', 'Finance', 'Marketing']
performance_ratings = [1, 2, 3, 4, 5]
attrition_status = ['Yes', 'No']
education_levels = ['High School', 'Bachelor\'s Degree', 'Master\'s Degree', 'PhD']
marital_statuses = ['Single', 'Married', 'Divorced', 'Widowed']
job_satisfaction_levels = [1, 2, 3, 4, 5]
work_locations = ['New York', 'San Francisco', 'Los Angeles', 'Chicago', 'Houston']

# Create a list to hold the data
data = []

# Generate the data
for _ in range(num_records):
    date_of_hire = fake.date_between(start_date='-10y', end_date='today')
    last_promotion_date = fake.date_between(start_date=date_of_hire, end_date='today')
    years_at_company = (pd.Timestamp.today() - pd.to_datetime(date_of_hire)).days // 365
    years_in_current_role = random.randint(0, years_at_company)
    years_since_last_promotion = (pd.Timestamp.today() - pd.to_datetime(last_promotion_date)).days // 365
    years_with_current_manager = random.randint(0, years_at_company)

    # Determine termination date if the employee has left
    if random.choice([True, False]):
        terminate_date = fake.date_between(start_date=date_of_hire, end_date='today')
    else:
        terminate_date = None

    record = {
        'Employee ID': fake.unique.random_number(digits=5, fix_len=True),
        'First Name': fake.first_name(),
        'Last Name': fake.last_name(),
        'Gender': fake.random_element(elements=('M', 'F')),
        'Age': random.randint(20, 60),
        'Job Role': fake.random_element(elements=job_roles),
        'Department': fake.random_element(elements=departments),
        'Salary': random.randint(30000, 120000),
        'Date of Hire': date_of_hire,
        'Performance Rating': fake.random_element(elements=performance_ratings),
        'Attrition': 'Yes' if terminate_date else 'No',
        'Education Level': fake.random_element(elements=education_levels),
        'Marital Status': fake.random_element(elements=marital_statuses),
        'Number of Projects': random.randint(1, 20),
        'Last Promotion Date': last_promotion_date,
        'Work Location': fake.random_element(elements=work_locations),
        'Training Hours': random.randint(0, 100),
        'Overtime': fake.random_element(elements=['Yes', 'No']),
        'Job Satisfaction': fake.random_element(elements=job_satisfaction_levels),
        'Monthly Income': random.randint(2500, 10000),
        'Years at Company': years_at_company,
        'Years in Current Role': years_in_current_role,
        'Years Since Last Promotion': years_since_last_promotion,
        'Years with Current Manager': years_with_current_manager,
        'Number of Training Sessions': random.randint(0, 20),
        'Terminate Date': terminate_date
    }
    data.append(record)

# Create a DataFrame
df = pd.DataFrame(data)


```

```python
# Save the DataFrame to a CSV file
df.to_csv('hr_analytics_data.csv', index=False)

print("HR Analytics data generated and saved to 'hr_analytics_data.csv'")
```
